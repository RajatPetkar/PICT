```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('nba.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Name'])
```

```python
df.isnull().sum()
```

```python
df['College'] = df['College'].fillna('Unavailable')
```

```python
df.isnull().sum()
```

```python
mean_Salary = df['Salary'].mean()
df['Salary'] = df['Salary'].fillna(value = mean_Salary)
```

```python
df.isnull().sum()
```

```python
data_desc = pd.DataFrame(df.groupby('Position').describe())
```

```python
data_desc['Salary']
```

```python
data_desc['Weight']
```

```python
import pandas as pd

summary_stats = df.groupby('Position')['Salary'].agg(
    mean_salary='mean',
    median_salary='median',
    min_salary='min',
    max_salary='max',
    std_salary='std'
)

print(summary_stats)
```

```python
bins = [15, 20, 25, 30, 35, 40]
labels = ['Youth', 'Young Adult', 'Prime', 'Experienced', 'Veteran']
# Youth = 15-20
# Young Adult = 21-25
# Prime = 26-30
# Expereinced=31-35
# Vetran=36-40
```

```python
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=bins,
    labels=labels,
    right=True,        
    include_lowest=True
)
```

```python
df.head()
```

```python
data_stats = df.groupby('AgeGroup').describe()
```

```python
data_stats['Salary']
```
