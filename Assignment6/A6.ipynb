```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Iris.csv")
```

```python

features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

plt.figure(figsize=(12, 8))

# Sepal Length
plt.subplot(2, 2, 1)
sns.histplot(df['SepalLengthCm'], kde=True, bins=20)
plt.title('Histogram of SepalLengthCm')

# Sepal Width
plt.subplot(2, 2, 2)
sns.histplot(df['SepalWidthCm'], kde=True, bins=20)
plt.title('Histogram of SepalWidthCm')

# Petal Length
plt.subplot(2, 2, 3)
sns.histplot(df['PetalLengthCm'], kde=True, bins=20)
plt.title('Histogram of PetalLengthCm')

# Petal Width
plt.subplot(2, 2, 4)
sns.histplot(df['PetalWidthCm'], kde=True, bins=20)
plt.title('Histogram of PetalWidthCm')

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))

# Sepal Length
plt.subplot(2, 2, 1)
sns.boxplot(y=df['SepalLengthCm'])
plt.title('Box Plot of SepalLengthCm')

# Sepal Width
plt.subplot(2, 2, 2)
sns.boxplot(y=df['SepalWidthCm'])
plt.title('Box Plot of SepalWidthCm')

# Petal Length
plt.subplot(2, 2, 3)
sns.boxplot(y=df['PetalLengthCm'])
plt.title('Box Plot of PetalLengthCm')

# Petal Width
plt.subplot(2, 2, 4)
sns.boxplot(y=df['PetalWidthCm'])
plt.title('Box Plot of PetalWidthCm')

plt.tight_layout()
plt.show()
```
